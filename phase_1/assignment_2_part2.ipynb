## Variables
Variables are used to store information to be referenced and manipulated in a computer program.

Variables are given the name "variable" because they have the capability to store data that can be altered or modified.

### Identifiers
Identifiers are the names you supply for **variables**, types, functions, and labels in your program.

In Python, valid identifiers are names that you can use to uniquely identify variables, functions, classes, and other program elements. These names must follow certain rules. Here's an explanation and some examples of valid identifiers in the context of Python:

### Rules for Valid Identifiers in Python:

**Rule 1: Must Start with a Letter or Underscore**

Valid Examples:
- `school_name`
- `abc`
- `_private_variable`

Invalid Examples:
- `123abc` (starts with a digit)
- `!variable` (starts with a special character)

**Rule 2: Can Contain Letters, Digits, and Underscores**

Valid Examples:
- `total_score`
- `user_123`
- `variable_name`

Invalid Examples:
- `user-age` (contains a hyphen/minus sign)
- `my$variable` (contains a special character other than underscore)

**Rule 3: Case-Sensitive**

Python treats identifiers with different letter cases as distinct entities.

Valid Examples:
- `example_variable`
- `Example_Variable`

**Rule 4: Cannot Be a Reserved Keyword**

Python keywords are reserved words that have predefined meanings in the language and cannot be used as identifiers.

These reserved words cannot be used to name any variable or other program element.

| Reserved Words |          |         |          |        |
|----------------|----------|---------|----------|--------|
| False          | await    | else    | import   | pass   |
| None           | break    | except  | in       | raise  |
| True           | class    | finally | is       | return |
| and            | continue | for     | lambda   | try    |
| as             | def      | from    | nonlocal | while  |
| assert         | del      | global  | not      | with   |
| async          | elif     | if      | or       | yield  |

### Initializing Variables
In Python, variables are defined and initialized at the same time, meaning that when we create a variable, we assign a value to it immediately.

**Example**

```python
name = "John Doe"
```

In this case, `name` is the variable name, and it is assigned the value `"John Doe"`, which is a string.

After initialization, this variable acts as an identifier for the value `"John Doe"`. You can then use or call the variable to access its assigned value.


````python
# Run this cell

name = "John Doe"      # String
age = 30               # Integer
height = 5.9           # Float
is_student = True      # Boolean

# Printing the values of the variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
````

## Data Strucutre

A **data structure** is a way of organizing and storing data in a specific format so that it can be efficiently accessed, modified, and processed. It helps in managing and arranging data in a logical and systematic manner to perform various operations on it. Example saving Student records in tabular manner.

In modern languages, some data structures are considered data types because it is now so easy to create and manipulate them.

| Data Structure          | Mutability | Definition                              | Example                                  |
| ----------------------- | ---------- | --------------------------------------- | ---------------------------------------- |
| List (Data Structure)   | Mutable    | Ordered collection of items.            | [1, 2, 3], ["apple", "banana", "cherry"] |
| Tuple (Data Structure)  | Immutable  | Ordered, immutable collection of items. | (1, 2, 3), ("apple", "banana", "cherry") |
| String (Data Structure) | Immutable  | Sequence of characters.                 | "Hello", 'Python'                        |

Note:-
- Mutable objects are those that allow you to change their value or data in place without affecting the object's identity. (**can be changed**)
- Immutable objects are those that don't allow you to change their value or data in place. (**permanent**)

## Basic String Operations
[Visit w3schools](https://www.w3schools.com/python/python_ref_string.asp) for more information.

<h3>Easy-to-understand string functions</h3>

| String Function | Description                                                  | Example                                                          |
| --------------- | ------------------------------------------------------------ | ---------------------------------------------------------------- |
| capitalize()    | Converts the **first character to upper case**               | "hello".capitalize()  "Hello"                                   |
| lower()         | Converts a string into **lower case**                        | "Hello".lower()  "hello"                                        |
| upper()         | Converts a string into **upper case**                        | "Hello".upper()  "HELLO"                                        |
| count()         | Returns the **number of times a specified value occurs**     | "apple".count("p")  2                                           |
| replace()       | Returns a string with a specified **value replaced**         | "Hello, world!".replace("world", "Python")  "Hello, Python!"    |
| split()         | Splits the string at the specified separator, returns a **list** | "apple,banana,orange".split(",")  ['apple', 'banana', 'orange'] |

These functions can be applied to strings to modify and manipulate text data in Python.

These examples demonstrate how to apply string function in Python.

**Example**
```py
text = "hello, world! How's it going?"
print(text.capitalize())             # Output: "Hello, world! how's it going?"
print(text.lower())                  # Output: "hello, world! how's it going?"
print(text.upper())                  # Output: "HELLO, WORLD! HOW'S IT GOING?"
print(text.count("o"))               # Output: 4
print(text.replace("world", "Python"))  # Output: "hello, Python! How's it going?"
print(text.split())                  # Output: ['hello,', 'world!', "How's", 'it', 'going?']
```


````python
# Run this cell
text = "hello, world! How's it going?"
print(text.capitalize())                # Output: "Hello, world! how's it going?"
print(text.lower())                     # Output: "hello, world! how's it going?"
print(text.upper())                     # Output: "HELLO, WORLD! HOW'S IT GOING?"
print(text.count("o"))                  # Output: 4
print(text.replace("world", "Python"))  # Output: "hello, Python! How's it going?"
print(text.split())                     # Output: ['hello,', 'world!', "How's", 'it', 'going?']
````

## `type()` Function

`type()` is a function that returns the data type of the value passed to it.

These examples demonstrate how to use the `type()` function to determine the data types of different values in Python.

**Example 1 - Using `type()` with an Integer (`int`):**

```py
print(type(35))   # Output: <class 'int'>
```

**Example 2 - Using `type()` with a Floating-Point Number (`float`):**

```py
print(type(3.14)) # Output: <class 'float'>
```

**Example 3 - Using `type()` with a String (`str`):**

```py
print(type("Hello, world!")) # Output: <class 'str'>
```

**Example 4 - Using `type()` with a Boolean (`bool`):**

```py
print(type(True)) # Output: <class 'bool'>
```
