## Variables
Variables are used to store information to be referenced and manipulated in a computer program.

Variables are given the name "variable" because they have the capability to store data that can be altered or modified.

image.png

### Identifiers
Identifiers are the names you supply for **variables**, types, functions, and labels in your program.

In Python, valid identifiers are names that you can use to uniquely identify variables, functions, classes, and other program elements. These names must follow certain rules. Here's an explanation and some examples of valid identifiers in the context of Python:

### Rules for Valid Identifiers in Python:

**Rule 1: Must Start with a Letter or Underscore**

Valid Examples:
- `school_name`
- `abc`
- `_private_variable`

Invalid Examples:
- `123abc` (starts with a digit)
- `!variable` (starts with a special character)

**Rule 2: Can Contain Letters, Digits, and Underscores**

Valid Examples:
- `total_score`
- `user_123`
- `variable_name`

Invalid Examples:
- `user-age` (contains a hyphen/minus sign)
- `my$variable` (contains a special character other than underscore)

**Rule 3: Case-Sensitive**

Python treats identifiers with different letter cases as distinct entities.

Valid Examples:
- `example_variable`
- `Example_Variable`

**Rule 4: Cannot Be a Reserved Keyword**

Python keywords are reserved words that have predefined meanings in the language and cannot be used as identifiers.

These reserved words cannot be used to name any variable or other program element.

| Reserved Words |          |         |          |        |
|----------------|----------|---------|----------|--------|
| False          | await    | else    | import   | pass   |
| None           | break    | except  | in       | raise  |
| True           | class    | finally | is       | return |
| and            | continue | for     | lambda   | try    |
| as             | def      | from    | nonlocal | while  |
| assert         | del      | global  | not      | with   |
| async          | elif     | if      | or       | yield  |

### Initializing Variables
In Python, variables are defined and initialized at the same time, meaning that when we create a variable, we assign a value to it immediately.

**Example**

```python
name = "John Doe"
```

In this case, `name` is the variable name, and it is assigned the value `"John Doe"`, which is a string.

After initialization, this variable acts as an identifier for the value `"John Doe"`. You can then use or call the variable to access its assigned value.


````python
# Run this cell

name = "John Doe"      # String
age = 30               # Integer
height = 5.9           # Float
is_student = True      # Boolean

# Printing the values of the variables
print("Name:", name)
print("Age:", age)
print("Height:", height)
print("Is Student:", is_student)
````

### Assignment Operator and Compound Assignment Operators

In Python, the **assignment operator** (`=`) is used to assign a value to a variable. For example, `x = 5` assigns the value `5` to the variable `x`.

| Operator | Description           | Example    |
| -------- | --------------------- | ---------- |
| `=`      | Assign                | `x = 5`    |

**Compound assignment operators** are a shorter way to perform an operation and then assign the result back to the same variable. They combine an arithmetic operator with the assignment operator. Some common compound assignment operators include:

| Operator | Description           | Example    | Equivalent |
| -------- | --------------------- | ---------- | ---------- |
| `+=`     | Add and Assign        | `x += 3`   | `x = x + 3` |
| `-=`     | Subtract and Assign   | `x -= 2`   | `x = x - 2` |
| `*=`     | Multiply and Assign   | `x *= 4`   | `x = x * 4` |
| `/=`     | Divide and Assign     | `x /= 2`   | `x = x / 2` |
| `%=`     | Modulus and Assign    | `x %= 7`   | `x = x % 7` |
| `//=`    | Floor Division Assign | `x //= 3`  | `x = x // 3` |
| `**=`    | Exponent and Assign   | `x **= 2`  | `x = x ** 2` |


These compound assignment operators allow you to perform the operation and assignment in a single step, making your code more concise and readable.

**Example 1 - Addition Assignment:**
```python
# a = a + b
a = 5
b = 3
a += b
print("Updated value of 'a' after addition:", a)   # Output: 8
```
Equivalent statement: `a = a + b`
